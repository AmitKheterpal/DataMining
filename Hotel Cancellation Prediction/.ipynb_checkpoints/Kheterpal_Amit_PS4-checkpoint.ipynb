# 1. Problem Background and Motivation <a class="car" id="problem-background"></a>

### Predict booking cancellations for ABC Group of Hotels

>**What is the problem in Question?**<br>
> Every year, our hotel loses a significant amount of potential revenue due to customer cancellations. We created a tool to predict if a customer will cancel so that management can take the necessary steps to avoid losing revenue due to vacant rooms.
> It will also allow hotel chains to make necessary adjustments for customers in order to reduce the number of cancellations. <br><br>

>**Who will be our stakeholders?**<br>
> * **Hotel management**: Management is without a doubt one of our stakeholders. They plan to analyze data and implement changes to reduce the number of cancellations. They may also overbook if our model predicts a customer will cancel in order to avoid revenue loss.<br>

>**Why is it important?**<br>
> We can use this tool to predict whether a customer is likely to cancel. We can also include a dashboard with historical data to identify patterns such as publicizing specific hotel types, looking at year-over-year trends to identify time periods with peak cancelations, any specific demographics with a high cancelation rate, does cancelation have anything to do with a long waiting period, and whether previous customers are more likely to cancel or not.

```python
#########################
#Additional user defined functions 
#########################
'''
In this Function we are using inbuilt libraries to create a datafraem with all relevant performance measures
'''
def class_perf_measures(model,y_test,probab_threshold):
    # Create empty lists to store metric values created within loop
    TP = []
    TN = []
    FP = []
    FN = []
    recall = []
    precision = []
    F1 = []
    Accuracy = []
    AUC_ROC =[]    

    # Start loop
    for i in probab_threshold:

        # Create class assignments given threshold value


        y_test_pred_class = (model.predict_proba(X_test)[:,1] >= i).astype(int)

        # Append lists with metric values
        TP.append(metrics.confusion_matrix(y_test, y_test_pred_class)[1,1])
        TN.append(metrics.confusion_matrix(y_test, y_test_pred_class)[0,0])
        FP.append(metrics.confusion_matrix(y_test, y_test_pred_class)[0,1])
        FN.append(metrics.confusion_matrix(y_test, y_test_pred_class)[1,0])
        recall.append(metrics.recall_score(y_test, y_test_pred_class).round(3))
        precision.append(metrics.precision_score(y_test, y_test_pred_class,zero_division=0).round(3))
        F1.append(metrics.f1_score(y_test, y_test_pred_class).round(2))
        Accuracy.append(metrics.accuracy_score(y_test, y_test_pred_class).round(2))
        AUC_ROC.append(metrics.roc_auc_score(y_test, y_test_pred_class))

    # Create dataframe
    result = pd.DataFrame({"threshold":threshold,
                           "TP":TP,
                           "TN":TN,
                           "FP":FP,
                           "FN":FN,
                           "Precision":precision,
                           "Recall":recall,
                           "f1":F1,
                           "AUC_ROC":AUC_ROC,
                           "Accuracy": Accuracy
                          })

    # Let's look at our dataframe
    return result

def data_stats(df,feature_list):
    '''
    This function we are calculating basic stats for all numerical columns. 
    To indicate outlier presence we are chcking if std/mean is greater than 1.5. Also if skewness of these columns
    are outside +-3 range
    '''
    temp =  df[feature_list].describe().round(2).T
    temp['median'] = df[feature_list].median().round(2)
    temp['outlier_presence'] = np.where((temp['std']/temp['mean']>2.5)| (df[feature_list].skew(axis = 0)>3) |(df[feature_list].skew(axis = 0)<-3),'True','False')
    return temp

def target_by_cats(X):
    '''
    This function assumes the input dataframe (X) has a numerically defined target
    variable in the first column and that all remaining features are categorical
    '''
    global_mean = X.iloc[:,0].mean()
    target = X.columns[0]
    for col in X.columns[1:]:
        df_group = X.groupby(by=col)[target].agg(['mean','count']).round(2)
        df_group['diff'] = (df_group['mean'] - global_mean).round(2)
        display(df_group.sort_values('count',ascending =False))

def target_by_cats_percent(X):
    
    for col in X.columns[1:]:
        df_group = (X.groupby(by=col)[X.columns[0]].value_counts(normalize=True)*100)
        
        display(df_group)








```

### 3.3.1 Dealing with Outliers <a class="car" id="outlier-detection"></a>
* [Go to Data Prep](#data-prep)

```python
df['booking_changes'].value_counts().plot(kind ='bar')
```

```python
from scipy import stats

cnt3 = df.shape[0]
print("Number of records before outlier removals: "+str(cnt3))
df = df.reset_index()
rows_with_outliers =  np.where((np.abs(stats.zscore(df['adr']))>3) | ((np.abs(stats.zscore(df['booking_changes']))>3)) |
               ((np.abs(stats.zscore(df['previous_cancellations']))>3)) | ((np.abs(stats.zscore(df['required_car_parking_spaces']))>3))
                              | ((np.abs(stats.zscore(df['total_of_special_requests']))>3))| ((np.abs(stats.zscore(df['previous_bookings_not_canceled']))>3)))
# list(rows_with_outliers[0])
df.drop(list(rows_with_outliers[0]),axis = 0,inplace=True)
cnt4 = df.shape[0]
print("Number of records after outlier removals: "+str(cnt4))
```

**Action**: Outliers are removed by limiting z score to +-3. This led to removal of around 9000 rows and reduce outlier variation.

## 3.5 Feature Transformation <a class="car" id="feature-transform"></a><br>

> **Objective:** <br>
> reduce country set to top 5 observations.<br>
> Create dummies for categorical variables

* [Go to Data Prep](#data-prep)

```python

#taking only top 5 countries and renaming all other by 'others'
def country_set(value):
    country_set = set(['PRT','GBR','FRA','ESP','DEU'])
    input_set = set(value.strip().split(' '))
    
    if (country_set & input_set):
        return list(country_set & input_set)[0]
    else:
        return 'others'
df['country'] = df['country'].apply(country_set)

# print(df['country'].unique())

```

```python
# create dummies for categorical data
# categorical variables
cat_vars = ['hotel','distribution_channel','deposit_type','market_segment','customer_type','country']
# Loop over categorical vars and encode each one
for var in cat_vars:
    df[var] = df[var].astype("category")

df = pd.get_dummies(df,columns=cat_vars, prefix_sep='_', drop_first=False)
```

**final categorigal data:**<br>
'is_repeated_guest', 'got_desired_room','distribution_channel','deposit_type','market_segment','customer_type','country'

## 3.7 Data Normalization <a class="car" id="normalization"></a><br>
> **The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information**<br>
> We want to use the summary stats(std dev. and mean) from training data to normalize both training and testing data. 
* [Go to Data Prep](#data-prep)

```python
#We are going to use the Z score to normalize data for selected columns
variables_to_scale = ['lead_time','adr','previous_cancellations','booking_changes','total_of_special_requests']

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# X_train[variables_to_scale] = StandardScaler().fit_transform(X_train[variables_to_scale])
# X_test[variables_to_scale] = StandardScaler().fit_transform(X_test[variables_to_scale])


for i in variables_to_scale:

    means = X_train[i].mean()
    std = X_train[i].std()
    X_train_scaled[i] = X_train_scaled[i].apply(lambda x: (x - means ) / std)
    X_test_scaled[i] = X_test_scaled[i].apply(lambda x: (x - means ) / std)


```

```python
# Logistics Regression Model
a =  datetime.datetime.now()
lr_model = LogisticRegression()

# Parameters for grid search
parameters = {
    'penalty':['l2'],
    'C': np.logspace(-4,4,10),
    'solver': ['liblinear']
}

# Running Grid search for LR model
clf = GridSearchCV(lr_model,
                   param_grid = parameters,
                   scoring = 'accuracy',
                   cv= 10,
                   n_jobs=-1
                  )

LR_final = clf.fit(X_train_scaled,y_train.values.ravel())


y_test_pred_class_LR = LR_final.predict(X_test_scaled)

b = datetime.datetime.now()

print('time taken to run the LR model ' + str((b-a).total_seconds()))

print("accuracy score for test data is ",metrics.accuracy_score(y_test,y_test_pred_class_LR).round(2))
print("Accuracy score for training data is",LR_final.best_score_)

```

```python
print('Initial parameters: ', clf.best_params_)
y_test_pred_class_LR_train = LR_final.predict(X_train_scaled)
print('Train F1: ',metrics.f1_score(y_train,y_test_pred_class_LR_train).round(2))
print('Test F1: ', metrics.f1_score(y_test,y_test_pred_class_LR).round(2))
```

```python
#AUC Curve
y_pred_LR = LR_final.predict_proba(X_test_scaled)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_LR)
auc1 = metrics.roc_auc_score(y_test, y_pred_LR).round(2)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc1))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)

plt.show()

plt.close()
```
